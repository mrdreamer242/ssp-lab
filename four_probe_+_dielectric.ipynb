```python
#funtion for best fit line
import numpy as np
import matplotlib.pyplot as plt
import math

def bfit(xlist,ylist,x_lower,x_upper): # defining the best fit line function
    s_x = sum(xlist) # summing x and y lists
    s_y = sum(ylist)
    y_hat = []
  # initial values
    s_xx = 0
    s_xy = 0
    ss_xy = 0
  # error parameters
    ss_xx = 0
    sse = 0
    n = len(xlist)


    xmean = (s_x/n) # mean of X ------
    ymean = (s_y/n) # mean of Y ------ this dot exists on the best fit line
  # loops for lists 
    for i in xlist:
        s_xx = s_xx + i**2

    for i in range(n):
        xyp = xlist[i]*ylist[i]
        s_xy += xyp
        ss_xx += (xlist[i] - xmean)**2
    
        ss_xy += ((xlist[i] - xmean)*(ylist[i] - ymean))


    for i in range(n):
        c = ((ss_xy/ss_xx)*xlist[i] + ymean - (ss_xy/ss_xx)*xmean)
        y_hat.append(c)

    for i in range(n):
        sse += (ylist[i] - y_hat[i])**2

  # best fit line parameters
    m = ((n*s_xy) - (s_x*s_y))/((n*s_xx) - (s_x**2)) #slope
  
    c = ((s_y*s_xx) - (s_x*s_xy))/((n*s_xx) - (s_x**2)) #intercept
  
    x = [x_lower,x_upper] # defining lower and upper bounds
    y1 = (m*x[0]) + c # lines
    y2 = (m*x[1]) + c #
  
    y = [y1,y2]


  ##################
  #Error
    s_yx = (1/(n))*(sse)
    s_m = np.sqrt((s_yx)/ss_xx)
    s_c = (s_yx)*(1/n + xmean**2/ss_xx)
  
  # plotting 
    plt.xlabel("$ln(T-T_{max}$)")
    plt.ylabel("$ln (\dfrac{1}{k} - \dfrac{1}{k_{max}})$")
    plt.grid()
    plt.scatter(xlist,ylist)
    plt.plot(x,y,color='red',label='Linear fit')
    #plt.legend(loc='upper right')

  

#########

    plt.scatter(xmean,ymean,color='green',s=150)
  

    print("Slope (diffusitivity \u03B3)=", round(m,4))
    print("Intercept (-log c') =", round(c,4))
    print("Error in slope =",round(s_m, 4))
    print("Error in intercept =",s_c)
    print("Curie-Wein's constant =", round(math.exp(-c)/10**4, 4), "10^{4}")
    return 
```

## Hall Effect


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def best_fit_plot(x_l,y_l):   
 #storing size of the list in N
    N=len(x_l) 
    #lists to store the sum
    Sx=0
    Sy=0
    Sxx=0
    Sxy=0
#loop over N and calculate different sums
    for i in range(0,N):
     Sx+=x_l[i]
     Sy+=y_l[i]
     Sxy+=x_l[i]*y_l[i]
     Sxx+=x_l[i]**2
    Rlist=[]
#calculate slope and intercept
    c=(Sy*Sxx-Sx*Sxy)/(N*Sxx-Sx**2)
    m=(N*Sxy-Sx*Sy)/(N*Sxx-Sx**2)
 #list to store values of y=mx+c
    Yi=[]
#calculate the different values of y=mx+c for each value of x(independent quantity of the two lists taken) and store it in Yi
    for i in x_l:
      Y=m*i+c
      Yi.append(Y)
 
    for i in range(N):
      Rlist.append((y_l[i]-m*x_l[i]-c)**2)
    R=sum(Rlist)
 
#error part
 
    E=(R/(N-2))**0.5
    
    Del=(N*Sxx)-(Sx**2)
 
    Ec=E*(Sxx/Del)**0.5
    Em=E*(N/Del)**0.5
 
#print m, c and plot the graph between the independent list given and the corresponding y values
    print("slope of line=",m)
    print("intercept=",c)
 
    print("R=",R)
    print("E=",E)
    print("Error in m:",Em)
    print("Error in c:",Ec)

    Rh = m*0.543e-3/0.555
    print('Hall cefficient =',Rh)

    e = 1.6e-19

    rho = 1/(Rh*e)

    print('Carrirer cncentration =',rho)
 
    plt.scatter(x_l,y_l)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.plot(x_l,Yi,'r')
    plt.grid()
    return None
```
